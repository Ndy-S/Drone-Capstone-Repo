```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import firestore
import pandas as pd

# Load Credential Key
cred = credentials.Certificate('./drone-capstone-386903-d2b13d68a9d6.json')
firebase_admin.initialize_app(cred)

db = firestore.client()
```

```python
# Get user data
users_ref = db.collection('users')
docs = users_ref.get()

for doc in docs:
    print(f'Document ID: {doc.id}')
    print(f'Document data: {doc.to_dict()}')
    print('---')
```

```python
# Insert laundry data
excel_file = './laundry_data.xlsx'
df = pd.read_excel(excel_file)

reviews_ref = db.collection('laundry')

for index, row in df.iterrows():
    laundry_name = row['laundry_name']
    icon = row['icon']
    phone_number = row['international_phone_number']
    opening_hours = row['current_opening_hours']
    rating = row['rating']
    address = row['address']
    latitude = row['latitude']
    longitude = row['longitude']

    query = reviews_ref.where('laundry_name', '==', laundry_name).where('phone_number', '==', phone_number)
    matching_docs = query.get()

    if len(matching_docs) > 0:
        print(f'Data already exists for {laundry_name}, skipping insertion.')
        continue

    doc_ref = reviews_ref.document()
    doc_ref.set({
        'laundry_name': laundry_name,
        'icon': icon,
        'email': "N/A",
        'phone_number': phone_number,
        'opening_hours': opening_hours,
        'rating': rating,
        'address': address,
        'latitude': latitude,
        'longitude': longitude,
        'price': "N/A"
    })

    print(f'Data inserted for {laundry_name}')

print('Data insertion completed.')
```

```python
# Insert user data
excel_file = './reviews_data.xlsx'
df = pd.read_excel(excel_file)

users_ref = db.collection('users')

for index, row in df.iterrows():
    review_author = row['review_author']

    query = users_ref.where('username', '==', review_author)
    matching_docs = query.get()

    if len(matching_docs) > 0:
        print(f'Data already exists for {review_author}, skipping insertion.')
        continue

    doc_ref = users_ref.document()
    doc_ref.set({
        'username': review_author,
        'email': "N/A",
        'address': "N/A",
        'password': "000",
        'pref_loc': True,
        'pref_rating': True
    })

    print(f'Data inserted for user: {review_author}')

print('Data insertion completed.')
```

```python
# Insert review data
excel_file = './reviews_data.xlsx'
df = pd.read_excel(excel_file)

reviews_ref = db.collection('reviews')

for index, row in df.iterrows():
    laundry_name = row['laundry_name']
    review_author = row['review_author']
    review_rating = row['review_rating']
    review_text = row['review_text']
    sentiment_score = row['sentiment_score']

    laundry_query = db.collection('laundry').where('laundry_name', '==', laundry_name)
    matching_laundries = laundry_query.get()

    if len(matching_laundries) > 0:
        laundry_id = matching_laundries[0].id
    else:
        laundry_id = 'N/A'

    user_query = db.collection('users').where('username', '==', review_author)
    matching_users = user_query.get()

    if len(matching_users) > 0:
        user_id = matching_users[0].id
    else:
        user_id = 'N/A'

    doc_ref = reviews_ref.document()
    doc_ref.set({
        'laundry_id': laundry_id,
        'review_author_id': user_id,
        'review_rating': review_rating,
        'review_text': review_text,
        'sentiment_score': sentiment_score
    })

    print(f'Data inserted for laundry: {laundry_name}, author: {review_author}. Laundry ID: {laundry_id}, Author ID: {user_id}')

print('Data insertion completed.')
```
